```python
def log_transform(data, col):
    for colname in col:
        if colname in data.columns:
            if (data[colname] == 1.0).all():
                data[colname + '_log'] = np.log(data[colname]+1)
            else:
                data[colname + '_log'] = np.log(data[colname])
        else:
            print(f"Column '{colname}' not found in DataFrame.")
            print("Available columns:")
            print(data.columns)

# Replace 'Fuel_Type' with 'Engine Fuel Type', 'Transmission' with 'Transmission Type', 
# 'Owner_Type' with 'Make', 'Location' with 'Model', 'Kilometers_Driven' with 'Driven_Wheels', 
# and 'Price' with 'MSRP'
fig, axes = plt.subplots(3, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'Engine Fuel Type', data = data, color = 'blue', 
              order = data['Engine Fuel Type'].value_counts().index);
sns.countplot(ax = axes[0, 1], x = 'Transmission Type', data = data, color = 'blue', 
              order = data['Transmission Type'].value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'Make', data = data, color = 'blue', 
              order = data['Make'].value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'Model', data = data, color = 'blue', 
              order = data['Model'].head(20).value_counts().index);
sns.countplot(ax = axes[2, 0], x = 'Make', data = data, color = 'blue', 
              order = data['Make'].head(20).value_counts().index);
sns.countplot(ax = axes[2, 1], x = 'Model', data = data, color = 'blue', 
              order = data['Model'].head(20).value_counts().index);
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);

plt.figure(figsize=(13,17))
sns.pairplot(data=data.drop(['Driven_Wheels','MSRP'],axis=1))
plt.show()
```

```python

fig, axarr = plt.subplots(4, 2, figsize=(12, 18))
data.groupby('Model')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("Model Vs Price", fontsize=18)
data.groupby('Transmission Type')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title("Transmission Vs Price", fontsize=18)
data.groupby('Engine Fuel Type')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("Fuel_Type Vs Price", fontsize=18)
data.groupby('Make')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("Owner_Type Vs Price", fontsize=18)
data.groupby('Make')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title("Brand Vs Price", fontsize=18)
data.groupby('Model')['MSRP'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title("Model Vs Price", fontsize=18)
data.groupby('Number of Doors')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title("Seats Vs Price", fontsize=18)
data.groupby('Car_Age')['MSRP'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12)
axarr[3][1].set_title("Car_Age Vs Price", fontsize=18)
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```
